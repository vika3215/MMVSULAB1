```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
url='https://raw.githubusercontent.com/vika3215/MMVSULAB1/main/Krasnoyarsk.csv'
df=pd.read_csv(url)

years=np.array(df['Years']).reshape(-1, 1)
population=np.array(df['Population'])

poly_features = PolynomialFeatures(degree = 2, include_bias = False)
year_poly = poly_features.fit_transform(years)
model = LinearRegression()
model.fit(year_poly, population)
predicted_population = model.predict(year_poly)
plt.plot(years, population)
plt.plot(years, predicted_population)
plt.plot(years, population)
plt.xlabel('year')
plt.ylabel('population')
plt.title('Population of Krasnoyarsk since 1991')

plt.show()
```
